```python
import pandas as pd
data = pd.read_csv('datasets/NewYork/IncomeEduReg.csv')
print(data.shape)
```

```python
data1 = data.dropna(how = 'any')
print(data1.shape)
```

```python
data1['EduAboveHS(%)'] = 1-data1['LessThanHS']/data1['PopOver25']
data1.head()
```

```python
data2 = data1[['Zipcode','IncomePerCapita','EduAboveHS(%)']]
data2.head()
```

```python
data2.to_csv('datasets/NewYork/nyc_census.csv')
```

```python
data_list = pd.read_csv('datasets/NewYork/filtered_listings_NYC.csv')
data_list = data_list.iloc[:,1:]
data_list.head()
```

```python
data_list.rename({'zipcode':'Zipcode'},axis = 1, inplace = True)
data_list['Zipcode'] = pd.to_numeric(data_list['Zipcode'], errors='coerce')
data_list = data_list.dropna(subset = ['Zipcode'])
data_list['Zipcode'].astype(int)
```

```python
data_merged = data2.merge(data_list, on = 'Zipcode', how = 'outer')
data_merged.head()
```

```python
data_merged.shape
```

```python
data_merged.to_csv('datasets/NewYork/nyc_merged.csv')
```
